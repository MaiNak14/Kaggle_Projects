```python
#One hot encode all categorical variables
def label_encode_categorical(df):
    # Create a copy of the DataFrame to avoid modifying the original
    df_encoded = df.copy()

    # Iterate through columns and label encode categorical variables
    for col in df_encoded.columns:
        if df_encoded[col].dtype == 'O':  # Check if the column is of object type (categorical)
            label_encoder = LabelEncoder()
            df_encoded[col] = label_encoder.fit_transform(df_encoded[col])

    return df_encoded

# Apply label encoding to categorical variables
df_encoded = label_encode_categorical(df)

# Print the result
print(df_encoded)
```

```python
# drop missing values from encoded df. All missing values are in SalePrice column and must be removed in order for models to run
df_cleaned = df_encoded.dropna()

X = df_cleaned.drop('SalePrice', axis=1)
y = df_cleaned['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Define the parameter grid including alpha values and cv values
param_grid = {
    'alpha': [0.01, 0.05, 0.1, 1, 10]
}

# Create the Lasso model
lasso_model = Lasso()

# Perform grid search with cross-validation
grid_search = GridSearchCV(lasso_model, param_grid, cv=10)
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```
